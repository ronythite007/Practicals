## Implement a program for retrieval of documents using inverted files.

## Inverted Index

An inverted index is a data structure that maps terms (words or other unique identifiers) to the documents or locations where they appear. It's essentially a reversed index compared to a regular index that maps documents to their terms. Inverted indexes are crucial for efficient full-text searches in databases, search engines, and other applications dealing with large collections of text. 


**Core Concepts:**

Terms: The unique words or phrases within a set of documents. 

Documents: The individual pieces of content being indexed (e.g., web pages, articles, database records). 
Index: A data structure that maps terms to the documents they appear in. 

How it works:

1. Tokenization:
    The text content of each document is broken down into tokens (words, phrases, etc.). 
2. Index Creation:
   For each unique term, the index stores a list of document IDs (or other location identifiers) where that term appears. 
3. Search:
   When a user searches for a term, the index quickly retrieves the list of documents containing that term, enabling fast retrieval of    relevant content. 

```python
documents = {
    1: "the quick brown fox jumps over the lazy dog",
    2: "never jump over the lazy dog quickly",
    3: "brown dogs are quicker than foxes",
    4: "a quick brown dog outpaces a quick fox"
}

def build_inverted_index(docs):
    inverted_index = {}
    for doc_id, text in docs.items():
        
        # Tokenize and normalize (lowercase)
        words = text.lower().split()
        for word in words:
            if word not in inverted_index:
                inverted_index[word] = set()
            inverted_index[word].add(doc_id)
    return inverted_index
```

###  Tokenize documents (split text into words, lowercase)

```python
tokenized_docs = {}

for doc_id, text in documents.items():
    tokens = text.lower().split()  # lowercase and split on spaces
    tokenized_docs[doc_id] = tokens

tokenized_docs  
```

### Build the inverted index

```python
inverted_index = {}

for doc_id, tokens in tokenized_docs.items():
    for word in tokens:
        if word not in inverted_index:
            inverted_index[word] = set()
        inverted_index[word].add(doc_id)

for word, doc_ids in inverted_index.items():
    print(f"{word}: {sorted(doc_ids)}")
```

### Define a simple search function

```python
def search(inverted_index, query):
    query_terms = query.lower().split()
    if not query_terms:
        return set()
    
    result_docs = inverted_index.get(query_terms[0], set()).copy()
    
    for term in query_terms[1:]:
        result_docs &= inverted_index.get(term, set())
    
    return result_docs
```

```python
queries = [
    "quick brown",
    "lazy dog",
    "brown fox",
    "quickly",
    "fox jumps"
]

for q in queries:
    results = search(inverted_index, q)
    print(f"Query: '{q}'")
    if results:
        print("Documents found:", sorted(results))
    else:
        print("No documents found.")
    print()
```

```python
User = input("Enter the query : ")
search(inverted_index, User)
```
