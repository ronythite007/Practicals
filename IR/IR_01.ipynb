## Write a program for pre-processing of a text document such as stop word removal, stemming. 

Stop word removal

Stemming (spaCy uses lemmatization, which is more accurate and preferred)


```python
import spacy
# Load spaCy English language model
nlp = spacy.load("en_core_web_sm")
```

```python
# Input text document
text = "The quick brown foxes were jumping over the lazy dogs in the park."
```

```python
# Process the text with spaCy
doc = nlp(text)
```

## Tokenization

```python
for token in doc:
    print(token)
```

## Stop Words

```python
for token in doc:
    if token.is_stop:
        print(token)
```

## Lemmatization

```python
for token in doc:
    print(token.lemma_)
```

## Function to clean the text/document

```python
def preprocess_text(text):
    doc = nlp(text)
    cleaned_tokens = []
    for token in doc:
        if not token.is_stop and not token.is_punct:
            cleaned_tokens.append(token.lemma_.lower())

    return " ".join(cleaned_tokens)
```

```python
preprocess_text("Hello, I am Rohan Thite an Machine Learning Engg")
```

Stop words like "the", "were", "over", "in" are removed.

Lemmatization turns "foxes" into "fox", "jumping" into "jump", etc.

spaCy does lemmatization, not stemming (like NLTK's PorterStemmer). Lemmatization is usually better.
