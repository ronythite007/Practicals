```python
import pandas as pd
```

```python
df = pd.read_csv("retail.csv")
df
```

```python
df.drop("invoice_no",axis="columns",inplace=True)
```

```python
df
```

```python
region=["North","East","West","South"]
import numpy as np

df["region"]=np.random.choice(region,size=len(df))
```

```python
df
```

```python
df["revenue"]=df.quantity*df.price
```

```python
df
```

```python
category_sales = df.groupby("category")["revenue"].sum().reset_index().sort_values(by="revenue", ascending=False)
```

```python
category_sales
```

```python
region_sales = df.groupby("region")["revenue"].sum().reset_index()
region_sales
```

```python
import seaborn as sns
```

```python
sns.barplot(x="region",y="revenue",data=region_sales)
```
