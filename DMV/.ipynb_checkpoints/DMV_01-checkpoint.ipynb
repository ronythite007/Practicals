## Perform data cleaning operations, such as handling missing values, removing duplicates, or correcting inconsistencies. 

The sales dataset is clean and well-prepared for analysis. It contains no duplicate records, and all relevant fields have been properly validated. Data types are consistent across columns, and there are no missing or null values. This ensures reliable and accurate insights can be derived from the data without the need for additional preprocessing.

## Convert the data into a unified format, such as a common dataframe or data structure, to enable seamless analysis. 

```python
# Align columns if needed
common_columns = list(set(csv_data.columns) & set(excel_data.columns) & set(json_data.columns))

csv_data = csv_data[common_columns]
excel_data = excel_data[common_columns]
json_data = json_data[common_columns]

# Concatenate all into a single dataframe
all_sales_data = pd.concat([csv_data, excel_data, json_data], ignore_index=True)

# Final check
print(all_sales_data.info())
```

## Data Transformation: merging datasets, splitting columns, deriving new variables

will only derive new variable

```python
all_sales_data['Total_Sales'] = all_sales_data['Order_Quantity'] * all_sales_data['Unit_Price']
all_sales_data['Profit_Margin'] = all_sales_data['Profit'] / all_sales_data['Revenue']
```

```python
all_sales_data.Profit_Margin
```

```python
all_sales_data['Total_Sales'] 
```

## Analyze sales data: descriptive stats and aggregation

```python
# a) Descriptive statistics
all_sales_data.describe()
```

```python
# b) Aggregate total sales by Product Category

sales_by_category = all_sales_data.groupby('Product_Category')['Total_Sales'].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
#c) Average order value

average_order_value = all_sales_data.groupby('Product')['Total_Sales'].mean()
print(average_order_value)
```

```python
# d) Distribution of product categories (counts)

category_counts = all_sales_data['Product_Category'].value_counts()
print(category_counts)
```

## Create visualizations, such as bar plots, pie charts, or box plots, to represent the sales data and gain insights into sales trends, customer behavior, or product performance. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style='whitegrid')

# a) Bar plot: Total Sales by Product Category
plt.figure(figsize=(10,6))
sales_by_category.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Category')
plt.ylabel('Total Sales')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# b) Pie chart: Product Category distribution
plt.figure(figsize=(8,8))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Product Category Distribution')
plt.ylabel('')
plt.show()
```

```python
# c) Box plot: Profit Margin by Age Group
plt.figure(figsize=(10,6))
sns.boxplot(data=all_sales_data, x='Age_Group', y='Profit_Margin')
plt.title('Profit Margin Distribution by Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# d) Line plot: Monthly sales trend over time

# Convert 'Date' column to datetime
all_sales_data['Date'] = pd.to_datetime(all_sales_data['Date'], errors='coerce')

# Drop rows where 'Date' could not be parsed
all_sales_data = all_sales_data.dropna(subset=['Date'])

# Group by month end and sum total sales
monthly_sales = all_sales_data.groupby(pd.Grouper(key='Date', freq='ME'))['Total_Sales'].sum()

# Plot
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
