# Dataset Description: City Air Quality Data

This dataset contains air quality measurements collected from a specific city over a period of time. It includes various pollutant levels, Air Quality Index (AQI) values, and categorization of air quality.

## Columns

| Column Name  | Description                                                         | Data Type    |
|--------------|---------------------------------------------------------------------|--------------|
| City         | Name of the city where measurements were taken                     | String       |
| Date         | Date of the measurement (YYYY-MM-DD)                               | Date         |
| PM2.5        | Concentration of particulate matter smaller than 2.5 micrometers   | Numeric      |
| PM10         | Concentration of particulate matter smaller than 10 micrometers    | Numeric      |
| NO           | Nitric oxide concentration                                         | Numeric      |
| NO2          | Nitrogen dioxide concentration                                    | Numeric      |
| NOx          | Nitrogen oxides concentration                                     | Numeric      |
| NH3          | Ammonia concentration                                            | Numeric      |
| CO           | Carbon monoxide concentration                                   | Numeric      |
| SO2          | Sulfur dioxide concentration                                    | Numeric      |
| O3           | Ozone concentration                                             | Numeric      |
| Benzene      | Benzene concentration                                          | Numeric      |
| Toluene      | Toluene concentration                                          | Numeric      |
| Xylene       | Xylene concentration                                         | Numeric      |
| AQI          | Air Quality Index value summarizing overall air quality        | Numeric      |
| AQI_Bucket   | Categorical bucket representing AQI severity (e.g., Good, Moderate, Severe) | Categorical  |

## Notes

- **Date** is recorded in a daily format and can be used for time-series analysis.
- Pollutant columns represent concentrations measured typically in micrograms per cubic meter (µg/m³) or parts per million (ppm).
- `AQI` is a standardized index used to communicate how polluted the air is on any given day.
- `AQI_Bucket` categorizes the AQI value into meaningful health impact groups to facilitate interpretation.
- The dataset may contain missing values or outliers which need to be handled during analysis.
- This data supports monitoring of air pollution trends, identifying major pollutant contributors, and assessing health risks over time.

### 4. Create line plots or time series plots to visualize the overall AQI trend over time.

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['AQI'], color='navy', label='AQI')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Trend Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Plot individual pollutant levels (e.g., PM2.5, PM10, CO) on separate line plots to visualize their trends over time. 

```python
plt.figure(figsize=(14,10))

pollutants = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3']

for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(len(pollutants), 1, i)
    plt.plot(df['Date'], df[pollutant], label=pollutant)
    plt.ylabel(pollutant)
    plt.title(f'{pollutant} Levels Over Time')
    plt.grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

### 6. Use bar plots or stacked bar plots to compare the AQI values across different dates or time periods

```python
daily_avg_aqi = df.groupby('Date')['AQI'].mean().reset_index()

plt.figure(figsize=(12,6))
plt.bar(daily_avg_aqi['Date'], daily_avg_aqi['AQI'], color='teal')
plt.xlabel('Date')
plt.ylabel('Average AQI')
plt.title('Average AQI by Date')
plt.xticks(rotation=45)
plt.show()
```

### 7. Create box plots or violin plots to analyze the distribution of AQI values for different pollutant categories.

```python
print(df['AQI_Bucket'].unique())
```

```python
print(df['AQI_Bucket'].apply(type).value_counts())
```

```python
# Convert to string (this converts NaNs to 'nan' string)
df['AQI_Bucket'] = df['AQI_Bucket'].astype(str)

# Filter out rows where 'AQI_Bucket' is 'nan'
df_filtered = df[df['AQI_Bucket'] != 'nan']

# Sort categories alphabetically (or define a custom order if you want)
categories = sorted(df_filtered['AQI_Bucket'].unique())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='AQI_Bucket', y='AQI', data=df_filtered, palette='cool', order=categories)
plt.title('Distribution of AQI by AQI Bucket')
plt.xlabel('AQI Bucket')
plt.ylabel('AQI')
plt.show()
```

### 8. Use scatter plots or bubble charts to explore the relationship between AQI values and pollutant levels.

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(df['PM2.5'], df['AQI'], s=df['CO']*10, c=df['NO2'], cmap='plasma', alpha=0.6)
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.title('AQI vs PM2.5 (Bubble size = CO, Color = NO2)')
plt.colorbar(scatter, label='NO2')
plt.grid(True)
plt.show()
```

### Heatmap of Correlations Between Pollutants and AQI

```python
plt.figure(figsize=(12,8))
corr_matrix = df[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Pollutants and AQI')
plt.show()
```
