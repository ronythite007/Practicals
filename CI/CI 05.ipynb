# Load Balancing (Simple Explanation + Concept)

## 📌 What is Load Balancing?
Load balancing is a technique used in distributed systems to **distribute incoming client requests across multiple servers** so that:
- No single server gets overloaded
- System runs efficiently
- Response time is reduced
- Resources are used properly

---

## 🧠 Why we need Load Balancing?
If all requests go to one server:
- Server becomes slow or crashes
- Performance decreases

So we use load balancing to:
- Share workload
- Improve speed
- Increase reliability

---

## ⚙️ How it works (Basic Idea)
1. Clients send requests
2. Load balancer receives requests
3. It selects a server using an algorithm
4. Request is forwarded to that server
5. Server processes and responds

---

## 🔄 Common Load Balancing Algorithms

### 1. Round Robin
- Requests are assigned one by one in circular order
- Each server gets equal chance

**Example:**
R1 → Server1
R2 → Server2
R3 → Server3
R4 → Server1


---

### 2. Random Method
- Requests are assigned randomly to servers
- Simple but not always balanced

---

### 3. Least Connection (basic idea)
- Request goes to server with least active connections
- Used in real systems

---

## 🧾 Summary
Load balancing is a method to distribute client requests among multiple servers using different algorithms like Round Robin and Random to ensure efficient system performance and avoid overload.

```python
# Load Balancing Simulation

import random

# List of servers
servers = ["Server1", "Server2", "Server3"]

# Client requests
requests = ["R1", "R2", "R3", "R4", "R5", "R6"]

# ---------------- ROUND ROBIN ----------------
print("ROUND ROBIN LOAD BALANCING\n")

i = 0
for r in requests:
    server = servers[i % len(servers)]  # rotate servers
    print(r, "→", server)
    i += 1


# ---------------- RANDOM ----------------
print("\nRANDOM LOAD BALANCING\n")

for r in requests:
    server = random.choice(servers)  # random server selection
    print(r, "→", server)
```
