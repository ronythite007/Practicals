# Clonal Selection Algorithm (Short Note)

## 📌 Definition
Clonal Selection Algorithm is an optimization technique inspired by the **immune system** where good solutions are selected, cloned, and mutated to get better results.

---

## ⚙️ Steps
1. Generate random solutions  
2. Evaluate fitness  
3. Select best solutions  
4. Clone them  
5. Mutate clones  
6. Replace with better solutions  

---

## 📊 Example
Maximize:  
x²

Population:  
[2, 5, 8]

Fitness:  
4, 25, 64

Best selected:  
[8, 5]

After cloning & mutation:  
Improved solutions generated

---

## 🎯 Key Idea
Best solutions are **repeated (cloned)** and **slightly changed (mutated)** to find the optimal answer.

---

## ⚡ One-line
Select → Clone → Mutate → Improve

## Clonal Selection Algorithm

```python
# Clonal Selection Algorithm (Simplified Version)

import random

# ---------------- FITNESS FUNCTION ----------------
# Higher fitness = better solution
def fitness(x):
    return x * x   # example: maximize x^2


# ---------------- INITIAL POPULATION ----------------
population = [random.randint(0, 20) for _ in range(5)]

print("Initial Population:", population)


# ---------------- CLONAL SELECTION PROCESS ----------------
for generation in range(5):

    # STEP 1: Evaluate fitness of all individuals
    scored = [(x, fitness(x)) for x in population]

    # STEP 2: Sort based on fitness (descending)
    scored.sort(key=lambda x: x[1], reverse=True)

    # STEP 3: Select best individuals
    selected = [x[0] for x in scored[:3]]

    # STEP 4: Clone best individuals (replication)
    clones = []
    for s in selected:
        clones.extend([s] * 2)  # each selected is cloned twice

    # STEP 5: Mutation (small random change)
    mutated = [c + random.randint(-1, 1) for c in clones]

    # STEP 6: New population = best + mutated
    population = selected + mutated[:2]

    print(f"\nGeneration {generation+1}: {population}")
```
