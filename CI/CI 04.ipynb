# Fuzzy Set Operations and Relations

This notebook demonstrates:
1. Union of fuzzy sets (max rule)
2. Intersection of fuzzy sets (min rule)
3. Complement of fuzzy set (1 - membership)
4. Difference of fuzzy sets
5. Fuzzy relation using Cartesian product
6. Max-Min composition of fuzzy relations

## Define Fuzzy Sets

```python
# Fuzzy Sets (membership values between 0 and 1)

A = {"a": 0.2, "b": 0.5, "c": 0.8}
B = {"a": 0.6, "b": 0.3, "c": 0.7}

print("Set A:", A)
print("Set B:", B)
```

## Union, Intersection, Difference, Complement

```python
# UNION → max of membership values
union = {x: max(A[x], B[x]) for x in A}

# INTERSECTION → min of membership values
intersection = {x: min(A[x], B[x]) for x in A}

# DIFFERENCE (A - B) → A AND (NOT B)
difference = {x: min(A[x], 1 - B[x]) for x in A}

# COMPLEMENT → 1 - membership value
complement_A = {x: 1 - A[x] for x in A}

print("Union:", union)
print("Intersection:", intersection)
print("Difference (A-B):", difference)
print("Complement of A:", complement_A)
```

## Fuzzy Relation (Cartesian Product)

```python
# Another fuzzy sets for relation
X = {"x1": 0.2, "x2": 0.7}
Y = {"y1": 0.5, "y2": 0.8}

# Fuzzy relation using Cartesian product
# Rule: R(x,y) = min(X[x], Y[y])

R1 = {(i, j): min(X[i], Y[j]) for i in X for j in Y}

print("Fuzzy Relation R1 (X × Y):")
print(R1)
```

## Second Fuzzy Relation

```python
# Second relation for composition
Y2 = {"y1": 0.6, "y2": 0.4}
Z = {"z1": 0.9, "z2": 0.3}

# R2 = Y × Z
R2 = {(i, j): min(Y2[i], Z[j]) for i in Y2 for j in Z}

print("Fuzzy Relation R2 (Y × Z):")
print(R2)
```

## Max-Min Composition

```python
# MAX-MIN composition of R1 and R2

result = {}

for (i, k1) in R1:
    for (k2, j) in R2:

        # matching intermediate variable
        if k1 == k2:

            value = min(R1[(i, k1)], R2[(k2, j)])

            # take MAX for final result
            result[(i, j)] = max(result.get((i, j), 0), value)

print("Max-Min Composition Result:")
print(result)
```
