# 🧬 Distributed Evolutionary Algorithm (DEAP)

## 🎯 Goal
Find best value of `x` that maximizes:
f(x) = x²

---

## 🧠 Basic Idea
- Create random solutions
- Check which are good
- Keep best ones
- Improve them using:
  - Crossover (mix)
  - Mutation (small change)
- Repeat

---

## 🔄 Steps

1. **Initialize Population**
   - Example: [2, -5, 7, 1.5]

2. **Evaluate Fitness**
   - Apply function: x²  
   - Example: [4, 25, 49, 2.25]

3. **Selection**
   - Pick better values more often

4. **Crossover**
   - Mix two values  
   - Example: (7 + -5)/2 = 1

5. **Mutation**
   - Small random change  
   - Example: 1 → 1.3

6. **Repeat**
   - Process continues for many generations

---

## ⚡ Distributed Concept
- Normally: fitness runs one-by-one (slow)
- Distributed: runs in parallel (fast)

````python
# Simple & Clean DEAP Code with Proper Output

import random
from deap import base, creator, tools

# 1. Define fitness (maximize)
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()

# 2. Create individuals in range [-10, 10]
toolbox.register("attr", random.uniform, -10, 10)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr, 1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# 3. Fitness function (bounded)
def evaluate(ind):
    x = max(-10, min(10, ind[0]))   # keep value in range
    ind[0] = x
    return (x**2,)

toolbox.register("evaluate", evaluate)

# 4. Operators
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

# 5. Create population
pop = toolbox.population(n=10)

print("\n=== Evolution Start ===\n")

# 6. Run generations
for gen in range(1, 6):
    
    # Selection
    offspring = toolbox.select(pop, len(pop))
    offspring = list(map(toolbox.clone, offspring))

    # Crossover
    for c1, c2 in zip(offspring[::2], offspring[1::2]):
        toolbox.mate(c1, c2)
        del c1.fitness.values
        del c2.fitness.values

    # Mutation
    for ind in offspring:
        toolbox.mutate(ind)
        del ind.fitness.values

    # Evaluate
    for ind in offspring:
        ind.fitness.values = toolbox.evaluate(ind)

    pop[:] = offspring

    # Best individual
    best = tools.selBest(pop, 1)[0]

    # Proper formatted output
    print(f"Gen {gen:02d} | Best x = {best[0]:6.2f} | Fitness = {best.fitness.values[0]:8.2f}")

print("\n=== Evolution End ===")
````

## 🧠 Key Understanding

- Both **x = 10** and **x = -10** give same fitness:

- So algorithm finds **optimal solution quickly** (in Gen 1 itself)

- After that, it **switches between -10 and 10**
  👉 because both are equally best

---

## ⚡ Why it stabilizes early?

- Small search space [-10, 10]  
- Simple function (x²)  
- Easy to find maximum
