#### Implement Ant colony optimization by solving the Traveling salesman problem using python 

Problem statement- A salesman needs to visit a set of cities exactly once and return to the original city. The task is to find the shortest possible route that the salesman can take to visit all the cities and return to the starting city.

# How the ACO Works

## 1. Distance Matrix
- We define distances between cities using a matrix.
- Example: distance[i][j] = distance from city i to city j.

---

## 2. Initialize Parameters
- `num_ants` → number of ants (solutions per iteration)
- `num_iterations` → how many times algorithm runs
- `alpha` → importance of pheromone
- `beta` → importance of distance
- `evaporation` → reduces pheromone over time

---

## 3. Initialize Pheromone Matrix
- Start with equal pheromone on all paths.
- Means initially all paths are equally likely.

---

## 4. Each Ant Builds a Route
For every ant:
- Start from a random city
- Select next city based on probability:
  
  Probability depends on:
  - More pheromone → higher chance
  - Shorter distance → higher chance

- Keep adding cities until all are visited

---

## 5. Calculate Route Distance
- Add distances between consecutive cities
- Add distance to return to starting city

---

## 6. Store Best Route
- Compare all routes
- Keep track of shortest (best) route found so far

---

## 7. Pheromone Evaporation
- Reduce pheromone on all paths:
  
  pheromone = pheromone * (1 - evaporation)

- Prevents unlimited accumulation

---

## 8. Update Pheromone
- Add pheromone to paths used by ants:
  
  shorter route → more pheromone added

- Helps future ants choose better paths

---

## 9. Repeat
- Repeat steps for multiple iterations
- Over time, best path gets strongest pheromone

---

## 10. Final Output
- Print:
  - Best route
  - Shortest distance

---

## 🧾 In Short
- Ants try different paths
- Good paths get more pheromone
- Bad paths fade away
- Best path is found after iterations


```python
import numpy as np
import random

# Distance matrix (example: 5 cities)
dist = np.array([
    [0, 2, 9, 10, 7],
    [2, 0, 6, 4, 3],
    [9, 6, 0, 8, 5],
    [10, 4, 8, 0, 6],
    [7, 3, 5, 6, 0]
])

n = len(dist)  # number of cities

# Parameters
num_ants = 5
num_iterations = 50
alpha = 1       # importance of pheromone
beta = 2        # importance of distance
evaporation = 0.5

# Initialize pheromone matrix
pheromone = np.ones((n, n))

# Function to calculate route distance
def route_distance(route):
    total = 0
    for i in range(len(route) - 1):
        total += dist[route[i]][route[i+1]]
    total += dist[route[-1]][route[0]]  # return to start
    return total

# Main ACO loop
best_route = None
best_distance = float('inf')

for iteration in range(num_iterations):
    all_routes = []

    for ant in range(num_ants):
        route = [random.randint(0, n-1)]  # start from random city

        # Build route
        while len(route) < n:
            current = route[-1]
            probs = []

            for city in range(n):
                if city not in route:
                    # probability = pheromone^alpha * (1/distance)^beta
                    p = (pheromone[current][city] ** alpha) * ((1 / dist[current][city]) ** beta)
                    probs.append((city, p))

            # choose next city based on probability
            cities = [c for c, p in probs]
            weights = [p for c, p in probs]
            next_city = random.choices(cities, weights=weights)[0]

            route.append(next_city)

        all_routes.append(route)

        # update best route
        d = route_distance(route)
        if d < best_distance:
            best_distance = d
            best_route = route

    # Evaporation (reduce pheromone)
    pheromone *= (1 - evaporation)

    # Add new pheromone
    for route in all_routes:
        d = route_distance(route)
        for i in range(len(route) - 1):
            pheromone[route[i]][route[i+1]] += 1 / d

# Output
print("Best Route:", best_route)
print("Shortest Distance:", best_distance)
```
